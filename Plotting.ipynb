```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("additional/plotting/diamonds.csv")
df.head()
```

```python
sns.scatterplot(x="carat", y="price", data=df)
```

```python
sns.scatterplot(x="carat", y="price", data=df, alpha=0.25, edgecolor=None)
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x="carat", y="price", data=df, alpha=0.25, edgecolor=None)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["price"], kde=True, bins=75, edgecolor=None)
```

```python
sns.countplot(x="cut", data=df)
```

```python
plt.figure(figsize=(10,6))
order = ["Fair", "Ideal", "Good", "Very Good", "Premium"]
sns.boxplot(x="cut", y="price", data=df, order=order, palette="plasma")
```
